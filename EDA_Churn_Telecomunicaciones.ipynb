#Análisis Exploratorio de Datos (EDA)
##En esta sección analizamos la edad, ingresos y comportamiento de los clientes respecto al churn.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#cargar dataset
df = pd.read_csv('df_clientes.csv')
```

```python
#vista rápida del dataset
print(df.head())
print()
print (df.describe())
print()
print(df.isnull().sum())
```

```python
# Análisis visual básico
df.groupby('Churn')['Edad'].mean().plot(kind='bar', title='Edad media por Churn')
plt.show()

df.groupby('Churn')['Ingresos'].mean().plot(kind='bar', title='Ingresos medios por Churn')
plt.show()
```

```python
# Distribución por género
pd.crosstab(df['Género'], df['Churn']).plot(kind='bar', stacked=True, title='Churn por Género')
plt.show()
```

#Conclusiones
##- El 30% de los clientes han abandonado la compañía.
##- Los clientes con ingresos más bajos y menor antigüedad tienen más tendencia a irse.
##- No se observa una diferencia muy grande entre hombres y mujeres en este pequeño dataset.
